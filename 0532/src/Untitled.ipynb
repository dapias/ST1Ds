```julia
using PyPlot
```

```julia
include("initialconditions.jl")
```

```julia
N = 10000
z = rand(3, N)
```

```julia
z[:,1]
```

```julia
for i in 1:N
    z[:,i] = initcond(10.,2.)
end
```

```julia
plt[:hist](z[2,:], 30, normed = true)
x = collect(-5:0.01:5)
beta = 10.
gaussian = sqrt(beta)/sqrt(2.*pi)*exp(-beta*x.^2/2.)
plot(x,gaussian)
```

```julia
plt[:hist](z[3,:], 30, normed = true)
x = collect(-10:0.01:10)
s = (x -2.)
logistic = exp(-s)./((1. + exp(-s)).^2.)
plot(x,logistic)
```
