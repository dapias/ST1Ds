# Numerical analysis of ergodicity of 1D potentials thermostatted using the DD algorithm

## Harmonic Oscillator

```julia
using PyPlot
using PyCall
@pyimport matplotlib as mpl
mpl.style[:use]("classic")
```

```julia
file1 = "zsectionWWHvHO"
file2 = "psectionWWHvHO"
s1 = readdlm("../poincaredata/sectionsandtrajectories/$file1")
s2 = readdlm("../poincaredata/sectionsandtrajectories/$file2");
```

```julia
fig = plt[:figure](figsize=(12,6))   
#fig[:subplots_adjust](hspace=.3)

ax = fig[:add_subplot](1,2,1)                                                      
ticks = collect(-5.0:1.0:5.0)
ax[:set_yticks](ticks)
ax[:grid](which="both")
ax[:plot](s1[:,1], s1[:,2], "b.", ms=0.5, alpha = 0.7)

ax[:set_xlim](-3.0,3.0)
ax[:set_ylim](-4.,4.)
ax[:set_xlabel](L"$q$", fontsize = 18)
ax[:set_ylabel](L"$p$", fontsize = 18)
ax[:text](-2.5, 3.5, L"$\zeta = 0$", fontsize=15)

ax2 = fig[:add_subplot](1,2,2)                                                      
ticks = collect(-0.5:0.25:0.5)
ax2[:set_yticks](ticks)
ax2[:grid](which="both")
ax2[:plot](s2[:,1], s2[:,2], "r.", ms=0.5, alpha = 0.7)

ax2[:set_xlim](-3.5,3.5)
ax2[:set_ylim](-0.5,0.5)
ax2[:set_xlabel](L"$q$", fontsize = 18)
ax2[:set_ylabel](L"$\zeta$", fontsize = 18)
ax2[:text](-0.0, 0.4, L"$p = 0$", fontsize=15)

plt[:tight_layout]()

plt[:savefig]("../plots/$file1.png")
```
