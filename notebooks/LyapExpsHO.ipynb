```julia
using PyPlot
using PyCall
@pyimport matplotlib as mpl
mpl.style[:use]("ggplot")
```

```julia
figname = "lyap10000HO"
```

```julia
l = readdlm("../data/HO/$figname");
```

```julia
fig = plt[:figure](figsize=(8,8))
fig[:subplots_adjust](hspace=.5)


ax = fig[:add_subplot](311)
ax[:set_xlabel](L"$\lambda_1$",fontsize="18")
ax[:set_ylabel](L"$f(\lambda_1)$",fontsize="18")
ax[:set_axis_bgcolor]("white")
ax[:hist](l[:,1], 30, normed = 1)
ax[:grid](which="both", color = "gray", linestyle = "dashed")
ax[:tick_params](axis="both", labelsize=14)



ax = fig[:add_subplot](312)
ax[:set_xlabel](L"$\lambda_2$",fontsize="18")
ax[:set_ylabel](L"$f(\lambda_2)$", fontsize="18")
ax[:hist](l[:,2], 30, normed = true)
ax[:set_xlim](-1.5e-4,1.0e-4)
ax[:set_axis_bgcolor]("white")
ax[:grid](which="both", color = "gray", linestyle = "dashed")
ax[:tick_params](axis="both", labelsize=14)
ax[:ticklabel_format](style="sci", axis="x" , scilimits = (0,0))

ax = fig[:add_subplot](313)
ax[:set_xlabel](L"$\lambda_3$",fontsize="18")
ax[:set_ylabel](L"$f(\lambda_3)$",fontsize="18")
ax[:set_axis_bgcolor]("white")
ax[:grid](which="both", color = "gray", linestyle = "dashed")
ax[:tick_params](axis="both", labelsize=14)
ax[:hist](l[:,3], 30, normed = 1)


plt[:tight_layout]()


plt[:savefig]("../plots/$figname.pdf")
```
