### Hellinger Distance

```julia
filename = "integralig3cHO"
data = readdlm("../hellinger/$filename");
```

```julia
t = data[:,1]
h = data[:,2];
```

```julia
filename = "integralX4v7MH"
data = readdlm("../hellinger/$filename");
t1 = data[:,1]
h1 = data[:,2];

t1 = vcat(data[:,1][1:4], data[:,1][6:15], data[:,1][17:end])
h1 = vcat(data[:,2][1:4], data[:,2][6:15], data[:,2][17:end]);
```

```julia
filename = "integral0F7YQP"
data = readdlm("../hellinger/$filename");
t2 = data[:,1]
h2 = data[:,2];

t2 = vcat(data[:,1][1:9], data[:,1][11:end])
h2 = vcat(data[:,2][1:9], data[:,2][11:end]);
```

```julia
fig = plt[:figure](figsize=(10,4))
#fig = plt[:figure]()

ax = fig[:add_subplot](1,2,1)


#ax[:set_xscale]("log")
#ax[:set_yscale]("log")

a = ax[:plot](t, h, "o--", ms = 5.0, label = "Harmonic" )
b = ax[:plot](t2, h2, "o--", ms = 5.0, label = "Quartic" )
#c = ax[:plot](t1, h1, "o--", ms = 5.0, label = "Mexican" )


ax[:set_xlim](0.0,1.3e6)
#ax[:set_ylim](10^-1.4, 10^-0.6)
ax[:set_xlabel](L"t", fontsize = 18)
ax[:set_ylabel](L"D_H", fontsize = 18)
ax[:ticklabel_format](style="sci", axis="x" , scilimits = (0,0))
#handles, labels = ax[:get_legend_handles_labels]()
#ax[:legend](handles, labels)
ax[:tick_params](axis="both", labelsize=16)


ax = fig[:add_subplot](1,2,2)
ax[:set_xscale]("log")
ax[:set_yscale]("log")

a, = ax[:plot](t, h, "o--", ms = 5.0, label = "Harmonic" )
b, = ax[:plot](t2, h2, "o--", ms = 5.0, label = "Quartic" )
#c, = ax[:plot](t1, h1, "o--", ms = 5.0, label = "Mexican" )

ax[:set_xlim](10^4.7,10^6.2)
ax[:set_ylim](10^-2.7, 10^-1.9)
ax[:set_xlabel](L"t", fontsize = 18)
ax[:set_ylabel](L"D_H", fontsize = 18)
#ax[:ticklabel_format](style="sci", axis="x" , scilimits = (0,0))
handles, labels = ax[:get_legend_handles_labels]()
#ax[:legend](handles, labels)
ticks = [10.^-1.9, 10^-2.2, 10^-2.5, 10^-2.7]
ax[:set_yticks](ticks)
ax[:get_yaxis]()[:get_major_formatter]()[:labelOnlyBase] = false

#ticks = [10.^5.0, 10^5.5, 10^6.0]
#ax[:set_xticks](ticks)
#ax[:get_xaxis]()[:get_major_formatter]()[:labelOnlyBase] = false
ax[:tick_params](axis="both", labelsize=16)

#fig[:legend]((a,b,c),labels,"upper right")
fig[:legend]((a,b),labels,"upper right")

plt[:tight_layout]()
plt[:savefig]("../hellinger/hellingerdistances.pdf")
```

### Kullback Leibler distance

```julia
filename = "integralig3cHO"
data = readdlm("../kullback/$filename");
```

```julia
t = data[:,1]
h = data[:,2];
```

```julia
filename = "integralX4v7MH"
data = readdlm("../kullback/$filename");
t1 = data[:,1]
h1 = data[:,2];

t1 = vcat(data[:,1][1:9], data[:,1][11:end])
h1 = vcat(data[:,2][1:9], data[:,2][11:end]);
```

```julia
filename = "integral0F7YQP"
data = readdlm("../kullback/$filename");
t2 = data[:,1]
h2 = data[:,2];

t2 = vcat(data[:,1][1:14], data[:,1][16:end])
h2 = vcat(data[:,2][1:14], data[:,2][16:end]);
```

```julia
fig = plt[:figure](figsize=(10,4))
#fig = plt[:figure]()

ax = fig[:add_subplot](1,2,1)


#ax[:set_xscale]("log")
#ax[:set_yscale]("log")

a = ax[:plot](t, h, "o--", ms = 5.0, label = "Harmonic" )
b = ax[:plot](t2, h2, "o--", ms = 5.0, label = "Quartic" )
c = ax[:plot](t1, h1, "o--", ms = 5.0, label = "Mexican" )


ax[:set_xlim](0.0,1.3e6)
#ax[:set_ylim](10^-1.4, 10^-0.6)
ax[:set_xlabel](L"t", fontsize = 18)
ax[:set_ylabel](L"D_{KL}", fontsize = 18)
ax[:ticklabel_format](style="sci", axis="x" , scilimits = (0,0))
#handles, labels = ax[:get_legend_handles_labels]()
#ax[:legend](handles, labels)

ax = fig[:add_subplot](1,2,2)
ax[:set_xscale]("log")
ax[:set_yscale]("log")

a, = ax[:plot](t, h, "o--", ms = 5.0, label = "Harmonic" )
b, = ax[:plot](t2, h2, "o--", ms = 5.0, label = "Quartic" )
c, = ax[:plot](t1, h1, "o--", ms = 5.0, label = "Mexican" )

ax[:set_xlim](10^4.7,10^6.2)
ax[:set_ylim](10^-2.7, 10^-1.7)
ax[:set_xlabel](L"t", fontsize = 18)
ax[:set_ylabel](L"D_{KL}", fontsize = 18)
#ax[:ticklabel_format](style="sci", axis="x" , scilimits = (0,0))
handles, labels = ax[:get_legend_handles_labels]()
#ax[:legend](handles, labels)
ticks = [10.^-1.7, 10^-2.1, 10^-2.4, 10^-2.7]
ax[:set_yticks](ticks)
ax[:get_yaxis]()[:get_major_formatter]()[:labelOnlyBase] = false

#ticks = [10.^5.0, 10^5.5, 10^6.0]
#ax[:set_xticks](ticks)
#ax[:get_xaxis]()[:get_major_formatter]()[:labelOnlyBase] = false
#ax[:tick_params](axis="both", labelsize=16)

fig[:legend]((a,b,c),labels,"upper right")

plt[:tight_layout]()
plt[:savefig]("../kullback/kullbackdistances.pdf")
```
